# line chart

```python
import numpy as np
import matplotlib.pyplot as plt
a=np.linspace(0,10,20)
b=np.linspace(10,30,20)
fig=plt.figure(figsize=(6,5))
ax1=plt.subplot()
ax1.plot(a)
ax1.plot(b)
```

```python
import numpy as np
import matplotlib.pyplot as plt
aa=np.linspace(0,10,20)
bb=np.linspace(10,30,20)
fig=plt.figure(figsize=(6,5))
ax11=plt.subplot(221)
ax22=plt.subplot(223)
ax11.plot(aa,linewidth=2.0,linestyle='-',color='g',alpha=0.2,marker='o')
ax22.plot(bb,linewidth=1.0,linestyle='--',color='r',alpha=1,marker='^')
ax11.set(title='a')
ax22.set(title='b')
```

# areaplot

```python
import numpy as np
import matplotlib.pyplot as plt
a=range(0,5)
b=[[1,2,3,4,5],[4,5,6,7,8],[9,6,7,8,10]]
fig=plt.figure(figsize=(4,3))
ax3=plt.subplot()
ax3.stackplot(a,b)
plt.title('areaplot')
plt.xlabel('a')
plt.ylabel('b')
```

# barplot

```python
import numpy as np
import matplotlib.pyplot as plt
data={'apple':10,'orange':30,'banana':40}
names=list(data.keys())
values=list(data.values())
fig=plt.figure(figsize=(6,8))
ax4=plt.subplot()
ax4.bar(names,values)
plt.title('barplot')
plt.xlabel('x')
plt.ylabel('y')
```

# histogram

```python
import numpy as np
import matplotlib.pyplot as plt
number=[12,34,56,78,98,678,456,789,908,234,56,45,667]
fig=plt.figure(figsize=(5,6))
ax5=plt.subplot()
ax5.hist(number, bins=7)
plt.show()
```

# using diamonds dataset in the visualizations below

```python
import seaborn as sns
data = sns.load_dataset("diamonds")
data.head()
```

# violin plot

```python
sns.violinplot(x="color", y="carat", data=data, palette="muted", color="royalblue", width=1)
#sns.violinplot(x="color", y="carat", data=data, width=1)
```

```python
sns.violinplot(x="color", y="carat", data=data, width=1)
```

```python
sns.kdeplot(data=data, x="carat", hue="cut")
#sns.kdeplot(data=data, x="carat",y=cut, hue="cut")
```
